## Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
df1=pd.read_csv("crime_data.csv")
df1
```

```python
df2=df1.rename(columns={'Unnamed: 0':'Place'})
df2
```

```python
#Normalise the data function
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(df2.iloc[:,1:])
df_norm
```

```python
#Elbow curve
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
df2['Clust']=x
df2
```

```python
df2.iloc[:,1:5].groupby(df2.Clust).mean()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df3=pd.read_csv("crime_data.csv")
df3
```

```python
df3.info()
```

```python
df=df3.iloc[:,1:5]
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```
